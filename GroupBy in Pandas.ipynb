```python
import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('googleplaystore.csv')

impute = SimpleImputer(missing_values = np.nan , strategy = 'mean')
impute.fit(df.iloc[ : , 2:3 ].values)
df.iloc[ : , 2:3 ] = impute.transform(df.iloc[ : , 2:3 ].values)

df = df.dropna()
df.head()
```

<h1>Q1. Name the apps that are in ART_AND_DESIGN with rating more then 4.5 in descending order WRT Ratings?</h1>

```python
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
df_pr = df_pr[df_pr['Rating']>4.5]

df_pr.sort_values(by = 'Rating' , ascending = False).head()
```

<h1>Q2. Which category is having maximum average rating Descending order?</h1>

```python
df.groupby('Category')['Rating'].mean().sort_values(ascending = False)
```

<h1>Q3. How many paid apps are there in each category in Descending Order?</h1>

```python
df_pr = df[df['Type']=='Paid']
df_pr.groupby('Category').count()['Type'].sort_values(ascending = False)
```
