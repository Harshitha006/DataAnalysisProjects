```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('googleplaystore.csv')

impute = SimpleImputer(missing_values = np.nan , strategy = 'mean')
impute.fit(df.iloc[ : , 2 : 3].values)
df.iloc[ : , 2 : 3] = impute.transform(df.iloc[ : , 2 : 3].values)

df = df.dropna()
df.head()
```

```python
df.isna().sum()
```

```python
df['Category'].unique()
```

<h1>1. How many free apps are there in ART_AND_DESIGN?</h1>

```python
df.head()
```

```python
df_values = df.values
df_values
```

```python
c = 0

for i in df_values : 
    if(i[1] == 'ART_AND_DESIGN' and i[6] == 'Free'):
        c+=1
print('There are ',c ,'free apps in ART_AND_DESIGN')
```

<h1>Q2. How many apps are there in ART_AND_DESIGN with rating more then 4.5?</h1>

```python
c = 0

for i in df_values : 
    if(i[1] == 'ART_AND_DESIGN' and i[3] > '4.5' ):
        c+=1
print('There are ',c ,' apps in ART_AND_DESIGN with rating more than 4.5')
```

<h1>Q3. How many apps are there in FAMILY with rating more then 4.5 and Free?</h1>

```python
c = 0

for i in df_values : 
    if(i[1] == 'FAMILY' and i[3] > '4.5' and i[6] == 'Free' ):
        c+=1
print('There are ',c ,' free apps for Family with rating more than 4.5')
```
